```python
def get_id_list_from_range_string(range_s):
  """
    given a range in str form of two positive ints, where:
       the first is < the second and 
       they are separated by a single hyphen
       examples: "11-22" or "998-1012"
    return a list of each int in this range AS A STR (including both edges)
    ex: "95-115" returns ["95", "96", ..., "114", "115"]
  """
  (r0, r1) = range_s.split("-")
  r0 = int(r0)
  r1 = int(r1)
  X = []
  i = 0
  while r0 + i <= r1:
    X.append( str(r0 + i))
    i += 1
  return X

print( get_id_list_from_range_string("95-115") )
```

```python
def get_range_strings_from_file(filename):
  # get a list of ranges from a file
  # returns a list of ranges in str form
  try:
    with open(filename, 'r') as f:
      content = f.read()
      #print(content)
      content = content.strip()
      ranges_list = content.split(",")
  except FileNotFoundError:
    print("no such file found")
  return ranges_list
  
Q2 = get_range_strings_from_file("day2_input.txt")
print(Q2)
for x in Q2:
  print(f"{x:30}:  {len(get_id_list_from_range_string(x))}")
```

```python
def is_invalid_as_double(X):
  """
  given an input str, looks if the word is the same word repeated twice
  so A A    2 2      2 2 2 2    2 3 2 3    1 2 3 1 2 3 are doubleLists and should return True
  and A B   1 1 1    2 3 2      2 3 3 2    1 2 3 1 2 3 1 2 3 are NOT doubleLists and should return False
  """
  # need at least two elements in list
  if len(X) < 2:
    return False
  
  # if odd number of elements, can't divide by 2
  if len(X) % 2 != 0:
    return False

  sublist_length = int(len(X) / 2)
  
  X1 = X[0: sublist_length]
  X2 = X[sublist_length : ]
  
  if X1 == X2:
    #print(f"invalid {X}")
    return True

  return False


test_cases = ["22", "2222", "2323", "111", "232", "2332", "123123123"]
for t in test_cases:
  #if type(t) == type(23):
  #  t = str(t)
  
  out = is_invalid_as_double(t)
  if out:
    print(f"{t:10} is a double code")
  else:
    print(f"{t:10} is not")
```

```python
def sum_valid_ids_in_range(range_str, valid_function):
  n = 0
  X = get_id_list_from_range_string(range_str)
  for x in X:
    if valid_function(x):
      n += int(x)
  return n
```

```python
def is_invalid_as_repeating(X, debug=0):
  """
  creating from is_invalid_as_double() - allow for list to repeat (fully) two or more times
  given an input string, is it the same chars repeated two or more times
  A A   A A A   A A A A   A B A B    A B C A B C would all match - be invalid
  A B   A B A   A B B A   A B C A B   would not match, not be invalid

  to do this we need to look at multiple lengths substrings
  max length substring is len(X) / 2
  only substr lengths that evenly divide len(X) are of interest
  """

  if len(X) < 2:
    return False

  if debug:
    print(f"Looking at {X}")
  sublist_length_to_try = 0
  while sublist_length_to_try <= len(X) / 2:
    # try dividing into equal parts each with sublist_length_to_try  length
    sublist_length_to_try += 1
    if debug:
      print(f"Checking lists of {sublist_length_to_try} items...")

    if len(X) % sublist_length_to_try != 0:
      #m += "not an even multiple"
      #print(m)
      continue

    if len(X) == sublist_length_to_try:
      continue
      
    
    num_sublists = int(len(X) / sublist_length_to_try)
    #m += f"in {num_sublists} parts"
    #print(m)
    
    # create new_list_of_sublists
    new_list_of_sublists = []
    for j in range(num_sublists):
      next_sublist = ""
      ##print(f"  Creating {j + 1} of {num_sublists}")
      for k in range(sublist_length_to_try):
        next_sublist +=   X[((j)*sublist_length_to_try) + k]
        ##print(next_sublist)

      ##print(f"  next_sublist: {next_sublist}")
      new_list_of_sublists.append(next_sublist)
      ##print(f"  checking {new_list_of_sublists}")

    # for this sublist_length_to_try we have new_list_of_sublists
    #print(f"sublist_length_to_try: {sublist_length_to_try}   new_list_of_sublists: {new_list_of_sublists}")

    # could do this recursively if we generalized but skipping for now
    k = 0
    any_diffs = False
    while k < len(new_list_of_sublists) - 1:
      if new_list_of_sublists[k] != new_list_of_sublists[k+1]:
        any_diffs = True
        #print(f" diff between {new_list_of_sublists[k]} and {new_list_of_sublists[k+1]}")
        break
      k += 1
    #print(f"end of while {k}")
    
    if not any_diffs:
      # then this one WORKS so return True...
      return True

  # eof - no trues, so must be false
  return False
      
```
