```python
def get_input_file(filename):
  try:
    with open(filename, 'r') as f:
      content = f.read()
      #print(content)
      content = content.strip()
      contents_as_list = content.split("\n")
  except FileNotFoundError:
    print("no such file found")
  return contents_as_list
```

```python
day3_ex = get_input_file("day3_example.txt")  
day3_in = get_input_file("day3_input.txt")  
```

```python
def find_max_joltageDEL(s, n):
  """
  given list of strings (0-9) find the n chars that make the highest joltage per the "joltage" rules
  s is one battery
  """


  if n > len(s):
    return 0

  # lets separate the first (len(x) - n) char from the final n char
  left_side = s[:-n]
  right_side = s[-n:]

  print(left_side)
  print(right_side)
  

  
  ## our base case is the last n digits
  best_joltage = s[ -n: ]

  return best_joltage


#find_max_joltage('987654321111111', 3)
#find_max_joltage('987654321', 3)



def get_largest_digitDEL(battery_s):
  """
  defined as the highest digit in the string - and if more than one the leftmost
  """

  maxVal = max(battery_s)
  maxVal_positions = []
  i = 0
  while i < len(battery_s):
    if battery_s[i] == maxVal:
      maxVal_positions.append(i)
    i += 1
  print(maxVal, maxVal_positions)

  return maxVal

  
def get_largest_avail_digit_and_update_battery(battery, spots, debug = 0):
  """
  joltage rules
  looking for highest digit in str
  for ties, leftmost digits are always preferred over rightmost digits
  need to reserve the last p positions for remaining spots

  """  
  if debug > 1:
    print(f"called for {spots} spots in: {battery}")

  joltage = ""
  while spots > 0:
    if debug > 0:
      print(f"tow: {spots} spots & joltage:>{joltage:10}<   battery:{battery}")
    
    # have to limit battery_avail so as to leave room for remaning spots if > 1
    spots_to_save = spots - 1
    p_term_string = len(battery) - spots_to_save
    battery_avail = battery[:p_term_string]
    
    if debug > 1:
      #print(f"need to save {spots_to_save} spots and term at {p_term_string}")
      print(f"battery_avail:                                {battery_avail}")
  
    maxVal = max(battery_avail)
    maxVal_positions = []
    i = 0
    while i < len(battery_avail):
      if battery_avail[i] == maxVal:
        maxVal_positions.append(i)
      i += 1

    
    p = maxVal_positions[0]
    q = len(battery) - p - 1
    # print(f"p {p}, q {q}")
    battery_revised = "0" * (p+1) + battery[-q: ]
    joltage += maxVal

    #print(f"battery:                                      {battery}")
    #print(f"battery_revised:                              {battery_revised}")
    battery = battery_revised
    spots -= 1

  return int(joltage)
    


n = 0
for x in day3_ex:
  z = get_largest_avail_digit_and_update_battery(x, 2)
  n += int(z)
  print(f"Adding {z} to get {n}")

```

```python
def maximize_joltage(batteries, flips, debug = 1):
  tot = 0
  for i, battery in enumerate(batteries):
    if debug:
      m = f"Checking battery {i+1:4}..."
    j = get_largest_avail_digit_and_update_battery(battery, flips, debug - 1)
    
    tot += j
    if debug:
      m += f" adding {j} to get {tot}"
      print(m)
  print(tot)   
```

```python
maximize_joltage(day3_ex, 2, 1)         
```

```python
maximize_joltage(day3_in, 2, 1)
```
