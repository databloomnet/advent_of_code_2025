```python
# output file
def get_input_file(filename):
  try:
    with open(filename, 'r') as f:
      content = f.read()
      contents_as_list = content.split("\n")
      if contents_as_list[-1] == "":
        contents_as_list = contents_as_list[:-1]
  except FileNotFoundError:
    print("no such file found")
  return contents_as_list

day7_example = get_input_file("day7_example.txt")
day7_example
```

```python
def find_all(s_in, x):
  # get list of idices where x is in s_in
  # to keep it easy, x must be str of one length and cannot be " "
  if len(x) != 1:
    print("error - need len one str")
    return []
  locs = []
  while True:
    p = s_in.find(x)
    if p == -1:
      break
    locs.append(p)
    s_in = s_in[0:p] + " " + s_in[p+1:]
  return locs

print( find_all(".^.^.^.^.^...^.", "^") )
print( find_all(".......S.......", "^") )
print( find_all(".......S.......", "S") )
```

```python
class TachyonManifold():
  def __init__(self, data):
    self.levels = []
    self.level_i_inc = 0
    self.level_i_nxt = 1
    
    #self.compute_next = 1
    #self.computed_levels = []

    self.total_splits = 0
    
    for d in data:
      self.levels.append( d )

    self.width = len(self.levels[0])


  def print(self):
    for v in self.levels:
      print(v)
    print(f"total splits: {self.total_splits}")

  def get_total_splits(self):
    return self.total_splits
  

  def compute(self):
    while self.level_i_nxt < len(self.levels):
      self.compute_one_step()
  
  def compute_one_step(self):
    debug = 0
    if debug > 2:
      print(f"tow level_i_inc: {self.level_i_inc}   level_i_nxt: {self.level_i_nxt}")
    assert self.level_i_inc + 1 == self.level_i_nxt, "level inc nxt misaligned"

    
    if self.level_i_nxt == 1:
      # special handling for first compute

      inc_level = self.levels[self.level_i_inc]
      nxt_level = self.levels[self.level_i_nxt]
      
      # no changes to the 0th level
      #self.computed_levels.append( inc_level )
      
      p = find_all(inc_level, "S")
      assert len(p) == 1, "error - expected to find a single S"

      nxt_level = nxt_level[0:p[0]] + "|" + nxt_level[p[0]+1:]
      #self.computed_levels.append( nxt_level )
      self.levels[self.level_i_nxt] = nxt_level
        
      self.level_i_inc += 1
      self.level_i_nxt += 1

      
    else:
      # start at level 3 or later (odds only)

      #assert self.level_i_inc % 2 == 0, "expected even level_i_inc"

      inc_level = self.levels[self.level_i_inc]
      nxt_level = self.levels[self.level_i_nxt]
      
      # scan incoming for beams
      beams =    find_all(inc_level, "|")

      # scan nxt for splitters
      splitters = find_all(nxt_level, "^")

      
      if debug > 1:
        print(f"beams: {beams}  splitters: {splitters}")
        print(f"inc_level:  {inc_level}")
        print(f"nxt_level:  {nxt_level}")
      
      
      # need to process every beam
      for b in beams:
        #print(f"beam {b}")
        if b in splitters:
          self.total_splits += 1
          # a beam hits a splitter, so need to propagate
          if b == 0:
            # special case - edge: only propagate right
            nxt_level = " |" + nxt_level[1:]
          elif b == len(inc_level) - 1:
            # special case - edge: only prop left
            nxt_level = nxt_level[0: b - 2] + "| "
          else:
            # regular scenario, prop both ways
            nxt_level = nxt_level[0: b - 1] + "|^|" + nxt_level[b+2:]
        else:
          # b is not in splitters, just carry down
          nxt_level = nxt_level[0: b] + "|" + nxt_level[b+1:]
        
        #print(f"--> {b}: {nxt_level}")

      # need to add ther new beams to the inc level

      self.levels[self.level_i_nxt] = nxt_level

      self.level_i_inc += 1
      self.level_i_nxt += 1      
      if debug > 0:
        print(f"nxt_level': {nxt_level}")

    return
      

X = TachyonManifold(day7_example)

X.compute()
X.print()
X.get_total_splits()
```
